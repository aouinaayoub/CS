```python
import math
import random
import matplotlib.pyplot as plt 

def cauchy(location, scale):
    p=0.0
    while p==0.0 :
        p = random.uniform(0, 1.0)
    return location + scale*math.tan(math.pi*(p - 0.5))
N=100
alpha = 30.0
beta= 10.0 
data_x= []
for i in range (N) :
    data_x.append(cauchy(alpha, beta)) 

plt.plot(range (N),data_x)
plt.xlabel('n')
plt.ylabel('x')
plt.show()


def cauchy_d(alpha, beta, x):
    return (1/(math.pi*beta))*(1/(1+((x-alpha)/beta)**2))
dist_data=[]
mean = []
mode = []
for alpha_unk in range (70):
    f=0 
    m=0
    norm=0
    for j in range (N):
        dist_data.append(cauchy_d(alpha_unk,beta, data_x[j]))
        m+=data_x[j]*cauchy_d(alpha_unk,beta, data_x[j]) 
        norm+=cauchy_d(alpha_unk,beta, data_x[j])
    mean.append(m/norm)
    mode.append(dist_data.index(max(dist_data)))
#print (f_data.index(max(f_data)))
#plt.plot(range(70),f_data)
print (m)
plt.plot (range(70),mode) 
plt.plot (range(70),mean)
plt.show()
```

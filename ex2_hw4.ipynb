```python
import numpy as np
import random 
import matplotlib.pyplot as plt 
L= 16
sigma=2
npins=9
x=[0]*npins
y=[0]*npins
a=0
m=True 

while m==True:
    a+=1
    for k in range (npins) :
        x[k]= random.uniform(sigma, L-sigma)
        y[k]= random.uniform(sigma, L-sigma) 
        m=False
        for i in range (k) :
            if (x[k]-x[i])**2+(y[k]-y[i])**2<=4*sigma**2 : 
                m=True
                break 
        if m==True : break 

print (x,y,a)
plt.scatter(x,y,marker='o',s=1000)
plt.show()

```

```python
import matplotlib.pyplot as plt 
def periodic_lattice(x,y) : 
    x=mod(x,L)
    if (x<0) : x=x+L
    y=mod(y,L)
    if (x<0) : y=y+L
    return(x,y)
from random import uniform as ran, choice

n=0
L=[(0,0)]*16*16
for i in range (16) :
    for j in range (16) : 
        L[n]=(0.5+i,0.5+j)
        n+=1

L_x=16      
sigma = 0.25; sigma_sq = sigma**2
delta = 0.5*sigma
for iter in range(100000):
    a = choice(L)
    L.remove(a)
    b = (a[0] + ran(-delta, delta), a[1] + ran(-delta, delta))
    min_dist = min((b[0] - x[0]) ** 2 + (b[1] - x[1]) ** 2 for x in L)
    lattice_cond = min(b[0],b[1]) < sigma or max(b[0], b[1]) > L_x - sigma
    if lattice_cond or min_dist < 4 * sigma ** 2:
        L.append(a)
    else:
        L.append(b)        

x_dat  = [0]*16*16 
y_dat = [0]*16*16

for i in range (16*16) : 
    x_dat[i]=L[i][0]
    y_dat[i]= L[i][1]
    
plt.scatter(x_dat,y_dat,marker='o',s=100)
plt.show()
print (L)
```

```python
import random                     
import matplotlib.pyplot as plt    
             
n_estimat= 1000
n_m=0
pi=3.14159265359
th_var_v= 4*pi-pi*pi
th_var= []
var_emp=[]   


for n_trials in range (50): 
    n_trials+=1
    s=0
    for j in range (n_estimat):
       # var_emp.append(j)
        n_m+=1
        n_inside = 0
        for i in range(n_trials):
            x, y = random.uniform(-1.0, 1.0), random.uniform(-1.0, 1.0)
            if x**2 + y**2 < 1.0: 
                n_inside += 1
        m= 4.0 * n_inside / float(n_trials)
        s+= (m - pi)*(m-pi) 
    th_var.append(th_var_v/n_trials)
    s/= n_estimat
    var_emp.append(s)
print (len(var_emp))
#plt.p
plt.plot(range (50),var_emp, label = 'experience')
plt.plot(range (50),th_var, label = 'Theor')
plt.legend(loc='lower left')
plt.xlabel('n')
plt.ylabel('variance')
plt.show()
```

```python
import math
import random
import matplotlib.pyplot as plt 

def cauchy(location, scale):
    p=0.0
    while p==0.0 :
        p = random.uniform(0, 1.0)
    return location + scale*math.tan(math.pi*(p - 0.5))
N=100
alpha = 30.0
beta= 10.0 
data_x= []
for i in range (10) :
    data_x.append(cauchy(alpha, beta)) 

plt.plot(range (10),data_x)
plt.xlabel('n')
plt.ylabel('x')
plt.show()


def cauchy_d(alpha, beta, x):
    return (1/(math.pi*beta))*(1/(1+((x-alpha)/beta)**2))
f_data=[]
for alpha_unk in range (70):
    f=0 
    for j in range (N):
        f+=math.log10(cauchy_d(alpha_unk,beta, data_x[j]))
    f_data.append(f)
print (f_data.index(max(f_data)))
plt.plot(range(70),f_data)
plt.plot (f_data.index(max(f_data)),max(f_data),'o') 
plt.show()
```

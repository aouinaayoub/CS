```python
import random
import math 
import matplotlib.pyplot as plt 
import numpy as np 
def exp_d(lamb,x):
    norm= lamb*(math.exp(-1/lamb)-math.exp(-20/lamb))
    return (math.exp(-x/lamb)/norm)
def exp_random(lamb,n):
    norm= lamb*(math.exp(-1/lamb)-math.exp(-20/lamb))
    data_x=[]
    for i in range (n) : 
        y=random.uniform( 0.0, lamb/float(norm) )
        x=lamb*math.log1p(1-y*norm/lamb)
        data_x.append(x)
    return (data_x)



n_data=1000
data_x= exp_random(10,n_data)

f_data =[]

for lamb_unk in range (1,50):
    
    f=0 
    for j in range (n_data):
        f+=math.log10(exp_d(lamb_unk,data_x[j]))
    f_data.append(f)
SE=[]
################ not sure
for i in range (1,50) : 
    SE.append((f_data.index(max(f_data)) - i)**2) 
plt.plot (range(1,50), f_data)
#plt.plot (range(1,50), SE)
plt.show()
```

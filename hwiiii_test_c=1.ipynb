```python

import random 
import math 
import matplotlib.pyplot as plt 

x_dat=[]
y_dat=[]
N=2300
for i in range (N):
    z=random.uniform(0.0, 1.0) 
    x=math.sqrt(z) 
    x_dat.append(x)
    px=2*x
    c=2
    y=random.uniform(0.0,c*px)
    y_dat.append(y)
    

plt.plot(x_dat,y_dat,'.')
plt.plot([0,1], [0,1])
plt.show()
```
